```python
from mistralai import Mistral

file_path = 'sample_data/benchmarks.csv'
with open(file_path, 'r') as file:
    data = file.read()

api_key = "<YOUR_API_KEY>"
model = "mistral-large-latest"

client = Mistral(api_key=api_key)

chat_response = client.chat.complete(
    model= model,
    messages = [
        {
            "role": "user",
            "content": (f"I have the following benchmark results attached as a string representing CSV file\n{data}\n"
            "Could you analyse all of the methods starting from T1 and provide short analytics about their performance."
            "Dont forget to compare them with each other."),
        },
    ]
)
print(chat_response.choices[0].message.content)


file_path = 'sample_data/code_samples.txt'

with open(file_path, 'r') as file:
    data = file.read()

chat_response = client.chat.complete(
    model= model,
    messages = [
        {
            "role": "user",
            "content": (f"I have the following C# code snippets\n{data}\n"
              "Could you please tell me if each of these snippets is performance efficient? If no - provide refactored snippets"),
        },
    ]
)
print(chat_response.choices[0].message.content)

```
